# Shopping List

Vom construi o pseudo-aplicație "listă de cumpărături" cu următoarele opțiuni active:
   - Ieșire din aplicație.
   - Afișare listă.
   - Adăugare de elemente.
   - Ștergere.
   - Sortare în ordine alfabetică.
   - Sortare în ordine inversă.
   
Această aplicație ne va ajuta să sedimentăm informațiile despre liste aflate din prezentare.

```python
# Initialize variables
# Start with an empty list of items
items = []
# ... and a choice equal to None
choice = None

# The application has to run while choice != 0
while choice != "0":
    # Before each new choice, print the 
    # options in the menu.
    print(
    """
    High Scores
    0 - Exit
    1 - Show Items
    2 - Add an Item
    3 - Delete an Item
    4 - Sort Items
    """
    )
    
    # Ask user to make a choice
    choice = input("Choice: ")
    # Print his/hers option
    print()

    # 0 - exit
    if choice == "0":
        print("Good-bye.")
    # 1 - list the items to buy, 1 per line
    elif choice == "1":
        print("Shopping list items")
        for item in items:
            print(item)
    # 2 - add an item at the end of the list
    elif choice == "2":
        item = input("What other item do you have to buy?: ")
        items.append(item)
    # 3 - remove an item only if it is in the list
    #     if not, inform the user
    elif choice == "3":
        item = input("Remove which item?: ")
        if item in items:
            items.remove(item)
        else:
            print(item, "isn't in the shopping list.")
    # 4 - sort items
    elif choice == "4":
        items.sort(reverse=True)
    # some unknown choice
    else:
        print("Sorry, but", choice, "isn't a valid choice.")

input("\n\nPress the enter key to exit.")

```
